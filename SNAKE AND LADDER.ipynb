```python
import random
position = [0,0]
snakes = {
    17:7,
    54:34,
    62:19,
    98:79
}
ladder = {
    3:38,
    24:33,
    42:93,
    72:84
}

print("......Welcome to snake and ladder game......")
player_name1 = input("Enter name of player 1:")
player_name2 = input("Enter name of player 2:")

print(".....Lets start the game.....")
Nam = 'Invalid Name'
while Nam == 'Invalid Name':
    print(" Welcome ! Who wants to begin the game?:")
    print(f"Enter {player_name1} or {player_name2}:")

    choice = input("Enter your choice:")

    Nam = 1 if choice==player_name1 else 2 if choice==player_name2 else "Invalid Name"
    if(Nam == 'Invalid Name'):
        print("Enter Correct Name")
    
print(choice, "starts the game") 

def start(i,names, position=[0,0] ):
    i = i-1
    while position[0]!=100 and position[1]!=100:
        
        roll = input(names[i]+" enter roll to roll or enter steps: ")
        if roll == "quit":
            print(names[((i+1)%2)], " wins the game")
            break
        elif roll == "roll":
            roll = random.randint(1,6)
            print("You got a ",roll)
        try:
            move = int(roll)
            if(move>20 or move<=0):
                print("Valid steps from 1 to 20 only") 
                i=(i+1)%2
            elif(position[i]+move<=100):
                position[i] = position[i] + move
            if position[i] == 100:
                print(f"Game finished {names[i]} wins")
                print("Your final position is", position[i])
                break
            if position[i] in ladder:
                print("Ladder it is")
                position[i] = ladder[position[i]]
            elif position[i] in snakes:
                print("Snake bite")
                position[i] = snakes[position[i]]
            print("Your final position is", position[i])
            i=(i+1)%2
        except:
            print("Invalid Choice")
            
start(Nam,(player_name1,player_name2), position)
```
